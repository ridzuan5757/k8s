```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import gridspec
```

```python
# Data
architecture = [1, 2, 3, 4, 5, 6, 7]
ap_southeast_1 = [1524.096, 1883.952, 1820.448, 1721.664, 1866.24, 1813.248, 1508.544]
ap_southeast_5 = [1410.048, 1745.28, 1583.28, 1535.328, 1624.536, 1617.84, 1299.888]
```

```python
def make_broken_bar_chart(ax1, ax2, x, data1, data2, labels):
    width = 0.35
    ax1.bar(x - width/2, data1, width, label=labels[0])
    ax1.bar(x + width/2, data2, width, label=labels[1])
    ax2.bar(x - width/2, data1, width, label=labels[0])
    ax2.bar(x + width/2, data2, width, label=labels[1])

    ax1.set_ylim(1000, 2000)
    ax2.set_ylim(0, 10)

    ax1.spines['bottom'].set_visible(False)
    ax2.spines['top'].set_visible(False)
    ax1.xaxis.tick_top()
    ax1.tick_params(labeltop=False)
    ax2.xaxis.tick_bottom()

    d = .015  # size of diagonal lines
    kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
    ax1.plot((-d, +d), (-d, +d), **kwargs)
    ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs)

    kwargs.update(transform=ax2.transAxes)
    ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)
    ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)
```

```python
fig = plt.figure()
gs = gridspec.GridSpec(2, 1, height_ratios=[1, 3])

# Create two axes
ax1 = fig.add_subplot(gs[0])
ax2 = fig.add_subplot(gs[1])

# Make the broken bar chart
make_broken_bar_chart(ax1, ax2, np.arange(len(architecture)), ap_southeast_1, ap_southeast_5, 
                      ['ap-southeast-1', 'ap-southeast-5'])

# Add labels and title
ax2.set_xlabel('Architecture')
ax2.set_ylabel('Value')
fig.suptitle('Comparison of ap-southeast-1 and ap-southeast-5', fontsize=16)
ax2.set_xticks(np.arange(len(architecture)))
ax2.set_xticklabels(architecture)

# Add legend
ax1.legend()

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
# Broken axis setup
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(8, 6))
fig.subplots_adjust(hspace=0.05)  # Adjust the space between plots

# Breakpoints for the y-axis
ax1.set_ylim(1700, 1900)  # Upper chart limit
ax2.set_ylim(1200, 1600)  # Lower chart limit

# Bar chart for ap_southeast_1
ax1.bar(architecture, ap_southeast_1, color='blue', label='ap_southeast_1')
ax2.bar(architecture, ap_southeast_1, color='blue')

# Bar chart for ap_southeast_5
ax1.bar(architecture, ap_southeast_5, color='orange', label='ap_southeast_5')
ax2.bar(architecture, ap_southeast_5, color='orange')

# Broken axis appearance
ax1.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax1.tick_params(labeltop=False)  # Don't put tick labels at the top
ax2.xaxis.tick_bottom()

# Add diagonal lines to indicate the break in the y-axis
d = .015  # size of diagonal lines
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # top-right diagonal

kwargs.update(transform=ax2.transAxes)  # Switch to the bottom axes
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # bottom-right diagonal

# Labels and title
ax2.set_xlabel('Architecture')
ax1.set_ylabel('Values')
ax2.set_ylabel('Values')
ax1.legend()

# Display the plot
plt.show()
```

```python
# Broken axis setup with side-by-side bars and adjusted y-axis break
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(8, 6))
fig.subplots_adjust(hspace=0.05)  # Adjust the space between plots

bar_width = 0.35  # Width of bars
index = np.arange(len(architecture))  # The x locations for the groups

# Breakpoints for the y-axis
ax1.set_ylim(1000, 2000)  # Upper chart limit
ax2.set_ylim(0, 1000)  # Lower chart limit

# Side-by-side bar chart
ax1.bar(index, ap_southeast_1, bar_width, color='blue', label='ap_southeast_1')
ax1.bar(index + bar_width, ap_southeast_5, bar_width, color='orange', label='ap_southeast_5')

ax2.bar(index, ap_southeast_1, bar_width, color='blue')
ax2.bar(index + bar_width, ap_southeast_5, bar_width, color='orange')

# Broken axis appearance
ax1.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax1.tick_params(labeltop=False)  # Don't put tick labels at the top
ax2.xaxis.tick_bottom()

# Add diagonal lines to indicate the break in the y-axis
d = .015  # size of diagonal lines
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # top-right diagonal

kwargs.update(transform=ax2.transAxes)  # Switch to the bottom axes
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # bottom-right diagonal

# Labels and title
ax2.set_xlabel('Architecture')
ax1.set_ylabel('Values')
ax2.set_ylabel('Values')
ax1.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
architecture = [1, 2, 3, 4, 5, 6, 7]
ap_southeast_1 = [1524.096, 1883.952, 1820.448, 1721.664, 1866.24, 1813.248, 1508.544]
ap_southeast_5 = [1410.048, 1745.28, 1583.28, 1535.328, 1624.536, 1617.84, 1299.888]

# Create figure and axis objects
fig, ax = plt.subplots()

# Create bars
x = np.arange(len(architecture))
width = 0.45  # default width

ax.bar(x - width/2, ap_southeast_1, width, label='ap-southeast-1 (Singapore)')
ax.bar(x + width/2, ap_southeast_5, width, label='ap-southeast-5 (Malaysia)')

# Set y-axis to start at 1000
ax.set_ylim(bottom=1000)

# Add labels and title
ax.set_xlabel('Architecture')
ax.set_ylabel('Monthly Cost / USD')
ax.set_title('Comparison of Monthly Cost Across Various Architecture')
ax.set_xticks(x)
ax.set_xticklabels(architecture)

# Add legend
ax.legend()

# Add a horizontal line at y=1000 to emphasize the cropping
ax.axhline(y=1000, color='r', linestyle='--', linewidth=0.8)

# Adjust layout and display the plot
plt.tight_layout()
plt.savefig("ec2.svg", format="svg")
plt.show()
```

```python
def add_arrays(arr1, arr2):
    result = []
    for i in range(len(arr1)):
        result.append(arr1[i] + arr2[i])
    return result
```

```python
vpc_1 = [1524.096, 1883.952, 1820.448, 1721.664, 1866.24, 1813.248, 1508.544]
vpc_5 = [1410.048, 1745.28, 1583.28, 1535.328, 1624.536, 1617.84, 1299.888]
ebs_1 = [457.54,494.40,494.40,610.75,531.26,616.42,760.05]
ebs_5 = [411.62,444.80,444.80,549.38,477.97,554.53,682.32]
total_1 = add_arrays(vpc_1,ebs_1)
total_5 = add_arrays(vpc_5,ebs_5)
```

```python
fig, ax = plt.subplots()

# Create bars
x = np.arange(len(architecture))
width = 0.45  # default width

ax.bar(x - width/2, total_1, width, label='ap-southeast-1 (Singapore)')
ax.bar(x + width/2, total_5, width, label='ap-southeast-5 (Malaysia)')

# Set y-axis to start at 1000
ax.set_ylim(bottom=1000)

# Add labels and title
ax.set_xlabel('Architecture')
ax.set_ylabel('Monthly Cost / USD')
ax.set_title('Comparison of Monthly Cost Across Various Architecture')
ax.set_xticks(x)
ax.set_xticklabels(architecture)

# Add legend
ax.legend()

# Add a horizontal line at y=1000 to emphasize the cropping
ax.axhline(y=1000, color='r', linestyle='--', linewidth=0.8)

# Adjust layout and display the plot
plt.tight_layout()
plt.savefig("vpc.svg", format="svg")
plt.show()
```
