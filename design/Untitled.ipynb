```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('pbi.xlsx', sheet_name="16.Supportability")
df = df.rename(columns=df.iloc[3])
df = df[["Col 4"]]
```

```python
def left_align(df: pd.DataFrame):
    left_aligned_df = df.style.set_properties(**{'text-align': 'left'})
    left_aligned_df = left_aligned_df.set_table_styles(
        [dict(selector='th', props=[('text-align', 'left')])]
    )
    return left_aligned_df
```

```python
pd.set_option('max_colwidth', 800)
df.assign(Name='need')
df["need"] = "no"
df.loc[[10, 11, 18, 20, 21, 26, 31, 32, 33, 37, 40, 49], "need"] = "yes"
```

```python
left_align(df)
```
